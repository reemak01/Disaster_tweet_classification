```python
train = pd.read_csv(LIBRARY_PATH + "preprocessed_data/clean_train.csv")
test = pd.read_csv(LIBRARY_PATH + "preprocessed_data/clean_test.csv")
```

## Bag of Words

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()

# Learn the vocabulary dictionary and return document-term matrix.
bag = vectorizer.fit_transform(train['text without sw_freq'].values)

# Split training data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(bag.toarray(), train['target'].values, test_size = 0.33, stratify= train['target'].values, random_state=30)
```

```python
print(vectorizer.get_feature_names_out())
print(vectorizer.vocabulary_)
print(bag.toarray())
```

```python
gnb = GaussianNB()
y_pred_gnb = gnb.fit(X_train, y_train).predict(X_val)
```

```python
print("Accuracy with validation set: ", (y_val == y_pred_gnb).sum() / X_val.shape[0])
```

## N-gram Bag of Words

#### Bigram

```python
vectorizer2 = CountVectorizer(ngram_range=(2, 2))
bag2 = vectorizer2.fit_transform(train['text without sw_freq'].values)
X_train2, X_val2, y_train2, y_val2 = train_test_split(bag2.toarray(), train['target'].values, test_size = 0.33, stratify= train['target'].values, random_state=30)
```

```python
gnb2 = GaussianNB()
y_pred_gnb2 = gnb2.fit(X_train2, y_train2).predict(X_val2)
print("Accuracy with validation set: ", (y_val2 == y_pred_gnb2).sum() / X_val2.shape[0])
```

#### Trigram

```python
vectorizer3 = CountVectorizer(ngram_range=(3, 3))
bag3 = vectorizer3.fit_transform(train['text without sw_freq'].values)
X_train3, X_val3, y_train3, y_val3 = train_test_split(bag3.toarray(), train['target'].values, test_size = 0.33, stratify= train['target'].values, random_state=30)
```

```python
gnb3 = GaussianNB()
y_pred_gnb3 = gnb3.fit(X_train3, y_train3).predict(X_val3)
print("Accuracy with validation set: ", (y_val3 == y_pred_gnb3).sum() / X_val3.shape[0])
```
